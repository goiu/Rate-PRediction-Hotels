```python
import pandas as pd 
pd.__version__
```

```python
business_json_path = '/Users/justenjoffe/Documents/yelp_academic_dataset_business.json'
business = pd.read_json(business_json_path, lines=True)
print(business.shape)
business.head()
```

```python
#Only keep businesses that are still open in the dataset
business.is_open.value_counts()
```

```python
df_business = business[business['is_open']==1].drop(['hours', 'is_open', 'review_count'], axis=1)
print(df_business.shape)
df_business.head()
```

```python
# Create one row for each series that contain comma-separated items
df_explode = df_business.assign(categories = df_business.categories.str.split(', ')).explode('categories')
df_explode.sample(3)
```

```python
#Split the businesses by category
print('Total number of categories:', len(df_explode.categories.value_counts()))
print('Top 30 categories:')
df_explode.categories.value_counts()[:30]
```

```python
#Find all categories under Hotels
df_explode[df_explode['categories'].str.contains('Hotels', case=True, na=False)].categories.value_counts()
```

```python
business_hotels= df_business[df_business['categories'].str.contains(
                         'Hotels', case=False, na=False)]
print(business_hotels.shape)
business_hotels.head()
```

```python
#Get the reviews
review_json_path = '/Users/justenjoffe/Documents/yelp_academic_dataset_review.json'
```

```python
#Use pandas to load large data by segmenting the file
size = 1000000
review = pd.read_json(review_json_path, lines=True,
                      # identifying the data type of each column can reduce memory usage
                      dtype={'review_id':str,'user_id':str,'business_id':str,'stars':int,
                             'date':str,'text':str,'useful':int,'funny':int,'cool':int},
                      chunksize=size)
```

```python
#Read one chunk of data at a time
chunk_list = []
for chunk in review:
    # Drop columns that aren't needed
    chunk = chunk.drop(['review_id','useful','funny','cool'], axis=1)
    # Renaming column name to avoid conflict with business overall star rating
    chunk = chunk.rename(columns={'stars': 'review_stars'})
    # Inner merge with edited business file so only reviews related to Fast Food remain
    chunk_merged = pd.merge(business_hotels, chunk, on='business_id', how='inner')
    # Show feedback on progress
    print(f"{chunk_merged.shape[0]} out of {size:,} related reviews")
    chunk_list.append(chunk_merged)
    
# After trimming down the review file, concatenate all relevant data back to one dataframe
df = pd.concat(chunk_list, ignore_index=True, join='outer', axis=0)
print(df.shape)
df.sample(3)
```

```python
#Convert the new data fram into a csv file creating more concise dataset
#Dataset further edited in csv form to remove further irrelevant information
df.to_csv("yelp_reviews_Hotels_categories.csv", index=False)
```
